```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Parâmetros do pêndulo e carrinho
carro_x = 0.0          # posição do carrinho (m)
carro_y = 0.07 + 0.07  # carrinho deslocado para ficar sobre as rodas
carro_largura = 0.4    # largura do carrinho (m)
carro_altura = 0.2     # altura do carrinho (m)
raio_roda = 0.07       # raio das rodas
comprimento_haste = 0.8  # comprimento do pêndulo (m)
angulo = np.deg2rad(0)  # ângulo do pêndulo em radianos

# Coordenadas da ponta do pêndulo
xp = carro_x + comprimento_haste * np.sin(angulo)
yp = carro_y + carro_altura + comprimento_haste * np.cos(angulo)

# Criação da figura
fig, ax = plt.subplots(figsize=(8,5))
ax.set_xlim(-2, 2)
ax.set_ylim(-0.2, 2)
ax.set_aspect('equal')
ax.set_title("Sistema de Pêndulo Invertido")

# Ativa a grade
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Desenha o chão
ax.plot([-3, 3], [0, 0], 'k', linewidth=2)

# Desenha carrinho (retângulo azul)
carrinho = patches.Rectangle((carro_x - carro_largura/2, carro_y),
                             carro_largura, carro_altura,
                             facecolor='blue')
ax.add_patch(carrinho)

# Desenha rodas (círculos pretos tocando o chão)
roda_esq = patches.Circle((carro_x - carro_largura/3, raio_roda), raio_roda, facecolor='black')
roda_dir = patches.Circle((carro_x + carro_largura/3, raio_roda), raio_roda, facecolor='black')
ax.add_patch(roda_esq)
ax.add_patch(roda_dir)

# Desenha a haste (linha preta)
ax.plot([carro_x, xp], [carro_y + carro_altura, yp], color='black', linewidth=2)

# Desenha a bola na ponta (vermelha)
bola = patches.Circle((xp, yp), 0.05, facecolor='red')
ax.add_patch(bola)

plt.show()
```
